```python
'''
IN GENERAL STRATEGY TO HANDLE MISSING VALUES:-

i) If a column has more than 80–90% missing values, it's usually not useful. so, it's safe to drop.

OTHERWISE:-

ii) handling data based on numerical columns:-
-if col = numerical:
if not too many outliers, fill na with mean of col.
else, fill na with median, as it is best used when data has too many outliers, and mean comes out to be false.

-if col = categorical:
fill na with mode of that column

-if col = datetime
fill na with placeholder value = '1900-01-01' in dtype = datetime

iii) drop rows if key columns in that row are missing. if key columns are very crucial, but they are missing, it is
safe to drop whole rows.

iv) verify that number of missing values are 0. it should be 0.

v) save cleaned data.
'''
```

```python
df.isnull().mean()*100
'''
here, i calculated the missing values as a percentage.
missing = df.isnull().sum
total = df.shape[0]

(missing/total) * 100
which is nothing but:-
mean*100
where mean is the average missing values.
'''
```

```python
df = df.loc[:, df.isnull().mean() < 0.85]

'''
diff bw loc and iloc.
loc = row label , col label.
used when you want to access row or col based on condition, and loc gives you those rows and cols which satisfy conditions

iloc = row index, col index
used when you want to access row or col based on index positions. conditions don't work here.

here, loc used, and all rows and only those col which has % of null values less than 85% are selected. rest dropped.
'''
```

```python
df.info()
```

```python
df.sample(10)
```

```python
numerical_cols = df.select_dtypes(include='number')
col_with_na = numerical_cols.columns[numerical_cols.isnull().any()]

'''
1st line = finding all columns that are numerical.
2nd line = finding only columns that are numerical as well as they have null values.

col_with_na stores only the names of the columns.
'''

col_with_na
```

```python
'''
if skewness is >1 or <-1, then the dist is NOT normal, and has many outliers.  under such condition, we use median. else
mean is used to fillna
'''
for i in col_with_na:
  #print(df[i].skew())
  if df[i].skew() > 1 or df[i].skew() < -1 :
    df[i].fillna(df[i].median(), inplace=True)
  else:
    df[i].fillna(df[i].mean(), inplace=True)

df.info()
```

```python
df.sample(20)
```

```python
obj_cols = df.select_dtypes(include='object')
obj_cols_with_na = obj_cols.columns[obj_cols.isnull().any()]

obj_cols_with_na

'''
finding all non numerical columns. it is NOT necessary that the non numerical columns we found are all categorical.
two types of non numerical col:-
1. categorical (check number of unique values which will often be very less, obviously much lesser than no. of rows)
2. free text (long text, no. of unique values will be quite much.)
3. dates, we always substitute 'missing' in place of missing values in dates, as we can't put any other values like mode
because it is very crucial info and wrong info can MISLEAD.
'''
```

```python
date_cols = ['Issued Date', 'Completed Date', 'First Construction Document Date', 'Permit Expiration Date']
for i in date_cols:
    df[i].fillna('Missing', inplace=True)
```

```python
df['Description'].fillna('No Description', inplace=True)

'''
if we use mode here, it can MISLEAD, as 2 diff entities can't have completely same description
'''
```

```python
cat_cols = [
    'Street Suffix', 'Existing Use', 'Proposed Use',
    'Existing Construction Type Description',
    'Proposed Construction Type Description',
    'Neighborhoods - Analysis Boundaries', 'Location'
]

'''
ONLY in case of categorical columns, we use mode to fill na or substitute with 'unknown'.
'''
for col in cat_cols:
    df[col].fillna('Unknown', inplace=True)
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
