```python
'''
there are 2 types of files.
1. csv (comma seperated values). so, when we use read_csv(), it seperates the features by commas and converts to a table format.
2. tsv (tab seperated values). so, when we use read_csv(), it can't separate the features because we don't have commas but we have tabs in between.


in general, if we see any features of the vector not seperated by the comma, it's not csv and when we read it, we don't get the table
format that we wanted. hence, for that, we use read_csv(sep=' ')  where between the inverted commas, we write that character which is separating
the values of the feature vectors, which obviously won't be a comma. if it were a comma, we would have NOT used sep=' '.

'''
```

```python
pd.read_csv('movie_titles_metadata.tsv')
```

### converters

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name

rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename})

'''
using converters, you can apply your defined function to any column. here, in team1 col, you want to rename every instance of
Royal Challengers Bangalore to RCB. so, you created the 'rename' func and simply applied it to the column
'''
```
