```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.head()
```

```python
temp_df=df.iloc[:,2:]
'''
gender is nominal, so ignored for now. we are right now just doing ordinal (review, education) and label (purchased) encoding.
'''
```

```python
temp_df.head()
```

```python
#ORDINAL ENCODING:-
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(temp_df.drop('purchased', axis=1),
                                                    temp_df['purchased'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape

'''
NOTE:-
we do train test split before applying any tranformations to the data. this is because we need only the train data to FIT. not the whole
data. after train data is fit, we apply transform to train as well as test data.

so, if you noticed in feature scaling, we first fit the X_train and then applying transform to both X_train and X_test.
similarly, here, as even this is a type of transform, we apply fit & transform to X_train and then only transform to X_test.
'''
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
#oe = OrdinalEncoder()
'''
just like label encoder below, i first passed just:-
oe = OrdinalEncoder()
I thought that, on doing oe.fit(X_train), we would get the ranks, as in poor = 0, avg = 1, and good = 2. but main thing is, computer is
a dumb machine and it does not know where poor, avg, and good stand. so, it randomly assigned the ranks, with poor being 2 and good being
1, and average being 0 which is a WRONG order.
hence, during ordinal encoding, specifying the order is important.
'''
```

```python
oe.fit(X_train)
```

```python
oe.categories_
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
X_train
```

```python
'''
label encoding is done only on the labels, not the cateogorical features.
'''
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
'''
as in the labels, there are no ranks and hence as order does not matter much, we don't write the order inside the le brackets.
'''
```

```python
le.fit(y_train)
```

```python
le.classes_
'''
as the name suggests, it has labels/classes and NOT categories. so sometimes, yes = 1 and no = 0, and sometimes, yes = 0 and no = 1.
'''
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
y_train
```
