```python
def separatestrings(str):
  a = str.strip().split()
  b=""
  c=""
  if len(a)==2:
    b=a[1]
    c=a[0]
    return c,b
  else:
    b=a[0]
    return c,b
separatestrings('A/5 2151')
```

```python
def separateStringsfromCabin(str):
  a=str[0]
  l = str.strip().split()
  b= l[0]
  c = ""
  for i in b:
    if i.isdigit():
      c=c+i

  return a,c

separateStringsfromCabin('C23 C25 C27')
```

```python
df[['ticket_cat', 'ticket_number']] = pd.DataFrame(df['Ticket'].apply(separatestrings).tolist())
```

```python
df.drop(columns=['cabin_num', 'cabin_cat'])
```

```python
df[['Cabin_cat', 'Cabin_number']] = pd.DataFrame(df['Cabin'].apply(separateStringsfromCabin).tolist())
'''
why did this error come?
biggest reason :- the df has NAN (missing values) in the Cabin column, and you are applying function to the Cabin column, assuming
everything is in strings. But, in pandas, the NAN are treated as float, and in function, you are wanting str. to error to ayega na!
therefore, it is said to remove or impute all the missing values.
ticket column me koi problem nahi aya, because there, no missing values present.
'''
```
