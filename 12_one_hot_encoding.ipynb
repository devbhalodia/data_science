### Why get into all this fuss? Simply use column transformer:-


```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer(transformers=[
    ('tnf1',OneHotEncoder(drop='first'),['fuel','owner','brand'])
],remainder='passthrough')
```

```python
df= pd.read_csv('cars.csv')
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['selling_price', 'km_driven']),df['selling_price'],
                                                test_size=0.2)
```
